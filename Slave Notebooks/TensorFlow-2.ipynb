```python
#to show relu and sigmoid functions plot

import matplotlib.pyplot as plt
import numpy as np

# Create an array of x-values
x = np.linspace(-5, 5, 100)  # Range from -5 to 5 with 100 points

# Define the ReLU and sigmoid functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Plot the ReLU and sigmoid functions
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, sigmoid(x), label='Sigmoid')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU and Sigmoid Activation Functions')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
