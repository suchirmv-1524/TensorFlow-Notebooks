```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons #binary classification dataset
from sklearn.model_selection import train_test_split
```

```python
x, y = make_moons(n_samples = 1000, noise = 0.1, random_state = 42) 
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(units=32, activation='relu', input_shape=x_train.shape[1:]), 
  tf.keras.layers.Dense(units=32, activation='relu'),
  tf.keras.layers.Dense(units=32, activation='relu'),
  tf.keras.layers.Dense(units=32, activation='relu'),
  tf.keras.layers.Dense(units=1, activation='sigmoid')  
])

model.compile(loss='binary_crossentropy',  
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```

```python
pd.DataFrame(history.history).plot()
```

```python
def plot_decision_boundary(model,X,y):
  # Define the axis boundaries of the plot and create a meshgrid
  x_min,x_max = X[:,0].min()- 0.1, X[:,0].max()+ 0.1
  y_min,y_max = X[:,1].min()- 0.1, X[:,1].max()+ 0.1
  xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                      np.linspace(y_min,y_max,100))
  #Create X values
  x_in = np.c_[xx.ravel(),yy.ravel()] #Stack 2-D drrays together
  # Make prediction:
  y_pred = model.predict(x_in)
  #Check for multi-class
  if len(y_pred[0])>1:
    print("Doing multi-class classification\n")
    # Reshape prediction for plotting
    y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("Doing binary classification\n")
    y_pred = np.round(y_pred).reshape(xx.shape)
  #Plot the decision boundary
  plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
  plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model, x_train, y_train)
```

```python
plot_decision_boundary(model, x_test, y_test)
```
