```python
def plot_decision_boundary(model,X,y):
  # Define the axis boundaries of the plot and create a meshgrid
  x_min,x_max = X[:,0].min()- 0.1, X[:,0].max()+ 0.1
  y_min,y_max = X[:,1].min()- 0.1, X[:,1].max()+ 0.1
  xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                      np.linspace(y_min,y_max,100))
  #Create X values
  x_in = np.c_[xx.ravel(),yy.ravel()] #Stack 2-D drrays together
  # Make prediction:
  y_pred = model.predict(x_in)
  #Check for multi-class
  if len(y_pred[0])>1:
    print("Doing multi-class classification\n")
    # Reshape prediction for plotting
    y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("Doing binary classification\n")
    y_pred = np.round(y_pred).reshape(xx.shape)
  #Plot the decision boundary
  plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
  plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model_5, x,y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2) 
#test size = 20% => train size = 80% => splitting into test and train data

model_7 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation = 'relu'), #relu is a non-linear curve
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid') #for binary classification, we'll have 1 neuron in output layer
    #when output layer has activation function as sigmoid, output gets pushed either to 0 or to 1
    #only output layer needs to necessarily have activation function
])

model_7.compile(loss = tf.keras.losses.BinaryCrossentropy(), optimizer = tf.keras.optimizers.Adam(), metrics = ['accuracy'])

history = model_7.fit(x_train,y_train,epochs = 100)

#for multiple classification into n classes, we'll have n neurons in output layer

model_7.predict(x)

#here, we get a very high accuracy due to non linear curves




```

```python
plot_decision_boundary(model_7, x_test,y_test)
#here, test data i.e. unseen data also is getting classified well => it's a good model
```

```python
import pandas as pd

pd.DataFrame(history.history).plot()

#history is the fitted data of previous data
# => plot shows that model is good with decreasing loss and increasing accuracy
```

```python
#finding the ideal learning rate so that there isn't any overshooting(not overfitting)
#overshooting means loss function will miss the ideal output everytime

model_8 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation = 'relu'), #relu is a non-linear curve
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid') #for binary classification, we'll have 1 neuron in output layer
    #when output layer has activation function as sigmoid, output gets pushed either to 0 or to 1
    #only output layer needs to necessarily have activation function
])

model_8.compile(loss = tf.keras.losses.BinaryCrossentropy(), 
                optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001), 
                metrics = ['accuracy'])

history_1 = model_8.fit(x_train, y_train, epochs = 100)

#learning rate = 1 => overshooting => horrible accuracy
#learning rate >= 1 => worst accuracy and worst loss
#learning rate <= 0.01 => decent loss and accuracy
#learning rate <= 0.001 => best loss and accuracy
#learning rate <= 0.0001 => bad loss and accuracy
```

```python
pd.DataFrame(history_1.history).plot()
```

```python
#to change learning rate for every epoch and find ideal learning rate

model_9 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation = 'relu'), #relu is a non-linear curve
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid') #for binary classification, we'll have 1 neuron in output layer
    #when output layer has activation function as sigmoid, output gets pushed either to 0 or to 1
    #only output layer needs to necessarily have activation function
])

lr = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4*10**(epoch/100))

model_9.compile(loss = tf.keras.losses.BinaryCrossentropy(), 
                optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001), 
                metrics = ['accuracy'])

history_2 = model_9.fit(x_train, y_train, epochs = 100, callbacks = [lr])


```

```python
pd.DataFrame(history_2.history).plot()
#learning rate slighly increases, and loss and accuracy remain the same
```
